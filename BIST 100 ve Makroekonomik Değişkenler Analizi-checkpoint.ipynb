```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error
```

```python
df = pd.read_excel("bist100_veriler.xlsx")
print(df.info())
print(df.describe())
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Korelasyon Matrisi")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error


X = df[["TÜFE (%)", "KUR (USD/TRY)", "FAİZ (%)", "İŞSİZLİK (%)"]]
y = df["BIST_KAPANIŞ"]

model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

print("R² Skoru:", r2_score(y, y_pred))
print("Ortalama Mutlak Hata (MAE):", mean_absolute_error(y, y_pred))
```

```python
plt.figure(figsize=(10,5))
plt.plot(df["YIL"], y, marker="o", label="Gerçek BIST")
plt.plot(df["YIL"], y_pred, marker="x", label="Tahmin Edilen BIST", linestyle="--")
plt.xlabel("Yıl")
plt.ylabel("BIST 100 Kapanış")
plt.title("Gerçek vs Tahmin Edilen BIST Kapanış Değerleri")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
